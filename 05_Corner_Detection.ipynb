# ♟ Chessboard Corner Detection – OpenCV

## Overview
This notebook is part of my **OpenCV learning roadmap**.  
The goal is to detect **corners** in an image (specifically a chessboard) using the **Shi-Tomasi corner detection algorithm** implemented in OpenCV.

## Objectives
- Load and preprocess an image by converting it to **grayscale**.
- Apply the `cv2.goodFeaturesToTrack()` function to identify strong corners.
- Visualize detected corners by marking them on the original image.

## Key Learnings
- Corner detection is useful in:
  - Camera calibration
  - Object tracking
  - Feature matching
- **Shi-Tomasi corner detection** finds points in an image where the surrounding pixels change in multiple directions.
- Key parameters in `cv2.goodFeaturesToTrack()`:
  - `maxCorners`: Maximum number of corners to detect.
  - `qualityLevel`: Minimum accepted quality of corners (0.01 means 1% of best corner quality).
  - `minDistance`: Minimum Euclidean distance between detected corners.
- Detected corners can be drawn using `cv2.circle()` for visualization.

## Tools & Libraries
- **OpenCV** – for image loading, grayscale conversion, corner detection, and drawing.
- **Matplotlib** – for displaying grayscale images inside Jupyter.
- **NumPy** – for coordinate manipulation.

---


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in grayscale
img = cv2.imread("./data/chessboard.png")
img = cv2.resize(img, (0, 0), fx=0.75, fy=0.75)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Chessboard")
plt.show()
```

```python
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 10)
corners = corners.astype(int)

for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 5, (255, 0, 0), -1)
        
cv2.imshow('Frame', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
